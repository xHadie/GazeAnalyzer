```python
class Subject:
    def __init__(self, name):
        name_parts = name.split("_")
        self.nationality = name_parts[0]
        self.name = name_parts[1]
        self.full_name = name
        
    def __str__(self):
        return "[%s - Nat.: %s]" % (self.name, self.nationality)
    

subjects_df = pd.read_csv(project_path + "subjects.csv")

def get_subject(name):
    for s in subjects_df["subject"]:
        if s == name:
            return Subject(s)

def get_all_subjects():
    return [Subject(s) for s in subjects_df["subject"]]

def get_subjects_of_nationality(nationality):
    return [Subject(s) for s in subjects_df["subject"] if s.startswith(nationality)]
    
    
```

```python
SMILE = "SMILE"
ANGRY = "ANGRY"
YAWN = "YAWN"

NUM_OF_INTENSITIES = 4
def number_to_emotion(num):
    if num == 1:
        return SMILE
    elif num == 2:
        return ANGRY
    elif num == 3:
        return YAWN
    
def intensity_of_image_index(image_index):
    return ((image_index - 1) % NUM_OF_INTENSITIES) + 1
    
```

```python
def gaze_string_tuple_to_float_tuple(tuple_str):
    tuple_str = tuple_str.replace("(", "")
    tuple_str = tuple_str.replace(")", "")
    tuple_str = tuple_str.replace(" ", "")
    values = tuple_str.split(',')
    if values[0] == "nan" or values[1] == "nan":
        return None, None
    return float(values[0]), (float(values[1]))


def create_big_gaze_data_file_for_subject(s: Subject):
    for session in [1,2,3]:
        gaze_data_path = data_path + s.full_name + "/%s" % session
        emotion = number_to_emotion(session)
        files = os.listdir(gaze_data_path)
        for f in files:
            # Gaze file has coding of et_imgNameIndex[img]_trial[trial_number].csv
            if f.startswith("et_imgNameIndex"):
                trial_path = gaze_data_path + "/" + f
                f = f.replace("et_imgNameIndex", "")
                f = f.replace(".csv", "")
                f = f.replace("trial", "")
                file_parts = f.split("_")
                
                # Meta Data
                image_number = int(file_parts[0])
                trial_number = int(file_parts[1])
                image_intensity = intensity_of_image_index(image_number)
                
                # Read gaze file and extract data
                print("Cond: %s - Img: %s (Intensity: %s) - Trial: %s" % (emotion, image_number, image_intensity, trial_number))
                trial_gaze_df = pd.read_csv(trial_path)
                
                for gaze_sample_index, row in trial_gaze_df.iterrows():
                    # Timestamps
                    os_timestamp = row["timestamp"]
                    eye_tracker_timestamp = row["ettimestamp"]
                    # Gaze location for left, right eyes and average of both
                    left_eye_x, left_eye_y = gaze_string_tuple_to_float_tuple(row["left_eye_gp"])
                    right_eye_x, right_eye_y = gaze_string_tuple_to_float_tuple(row["right_eye_gp"])
                    avg_eye_x = None
                    avg_eye_y = None
                    if left_eye_x is not None and right_eye_x is not None:
                        avg_eye_x = (left_eye_x + right_eye_x) / 2.0
                        avg_eye_y = (left_eye_y + right_eye_y) / 2.0

                    #print("AVG(%s, %s) | R(%s, %s), L(%s, %s) | T-OS: %s - T-ET: %s" % (avg_eye_x, avg_eye_y, right_eye_x, right_eye_y, left_eye_x, left_eye_y, os_timestamp, eye_tracker_timestamp))
        break
```
